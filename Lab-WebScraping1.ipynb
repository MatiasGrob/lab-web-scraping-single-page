```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# download the webpage
url = "https://www.popvortex.com/music/charts/top-100-songs.php"

# 3. download html with a get request
response = requests.get(url)
response.status_code # 200 status code means OK!

# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")

# print the formatted html
#soup
```

```python
# 5. retrieve/extract the desired info (here, you'll paste the "Selector" you copied before to get the element that belongs to the top movie)
#soup.select("body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8")
```

```python
soup.select("cite.title") # all the info about all the movies
```

```python
soup.select("em.artist") # all the info about all the movies
```

```python
#initialize empty lists
title = []
artist = []
genre = []

# define the number of iterations of our for loop
num_iter = len(soup.select("cite.title"))

title_list = soup.select("cite.title")
artist_list = soup.select("em.artist")
genre_list = soup.select('div.chart-content.col-xs-12.col-sm-8 > ul >li >a')

# iterate through the result set and retrive all the data
for i in range(num_iter):
    title.append(title_list[i].get_text())
    artist.append(artist_list[i].get_text())
    genre.append(genre_list[i+1].get_text())
```

```python
# each list becomes a column
movies = pd.DataFrame({"title":title,
                       "dir_stars":artist,
                       "Genre":genre
                      })
```

```python
# Scrape the current top 100 songs and their respective artists, and put the information into a pandas dataframe.
movies
```
